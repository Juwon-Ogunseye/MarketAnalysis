```python
import pandas as pd 
#read CSV file into a pandas data frame
df = pd.read_csv('/home/juwon/Documents/Final_finalsheet - Final_finalsheet.csv')
df.head(48)
```

```python
print(df.dtypes)
```

```python
df = df.dropna()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate the year-to-year changes for each column
df_yoy = (df.iloc[:, :-1].diff() / df.iloc[:, :-1].shift()) * 100

# Add the date column back to the DataFrame
df_yoy['date'] = df['date']

# Melt the DataFrame to long format
df_long = pd.melt(df_yoy, id_vars='date', var_name='country', value_name='yoy_change')

# Convert the date column to integer
df_long['date'] = df_long['date'].astype(int)

# Pivot the data to create a stacked bar chart
df_pivot = df_long.pivot_table(index='country', columns='date', values='yoy_change')

# Plot the stacked bar chart
sns.set_style('whitegrid')
plt.figure(figsize=(12, 8))
ax = df_pivot.plot(kind='bar', stacked=True, width=0.8, cmap='coolwarm')
plt.title('Year-to-Year Changes by Country')
plt.xlabel('Country')
plt.ylabel('Year-to-Year Change (%)')
plt.xticks(rotation=90)
plt.legend(title='Year', loc='upper left')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate the year-to-year changes for each column
df_yoy = (df.iloc[:, :-1].diff() / df.iloc[:, :-1].shift()) * 100

# Add the date column back to the DataFrame
df_yoy['date'] = df['date']

# Melt the DataFrame to long format
df_long = pd.melt(df_yoy, id_vars='date', var_name='country', value_name='yoy_change')

# Convert the date column to integer
df_long['date'] = df_long['date'].astype(int)

# Create a clustered column chart
sns.set_style('whitegrid')
g = sns.catplot(data=df_long, x='country', y='yoy_change', hue='date', kind='bar', col='date', col_wrap=4, dodge=False, palette=['#6d904f','#fc4f30','#008080','#af7ac5','#fbeb50','#a0cbe8','#fe9b87','#ff796c','#cedfd9','#d0e0eb','#ffa07a'])
g.fig.subplots_adjust(top=0.9)
g.fig.suptitle('Year-to-Year Changes by Country')
g.set_axis_labels('Country', 'Year-to-Year Change (%)')
g.set_xticklabels(rotation=90)
g.add_legend(title='Year', bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
plt.show()
```
